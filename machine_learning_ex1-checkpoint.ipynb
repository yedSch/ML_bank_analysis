###18

```python
from sklearn.utils import resample

X_res, Y_res = resample(X,y)
X_train_res, X_test_res, Y_train_res, Y_test_res = train_test_split(X_res, Y_res, test_size=0.3)
clf_gini = tree.DecisionTreeClassifier(max_depth=3, min_samples_leaf=2)
clf_gini.fit(X_train_res,Y_train_res)

tree.plot_tree(clf_gini)

y_pred_resample = clf_gini.predict(X_test_res)
print ('accuracy_resample',accuracy_score(Y_test_res, y_pred_resample)*100)


y_hat_probs_resample = clf_gini.predict_proba(X_test_res)[:, 1:]
print('auc_resample',roc_auc_score(Y_test_res, y_hat_probs_resample)*100)

confusion_matrix(Y_test_res, y_pred_resample)
```

###19

```python
clf_entropy = tree.DecisionTreeClassifier(max_depth=3, min_samples_leaf=2)
clf_entropy.fit(X_train_res,Y_train_res)

tree.plot_tree(clf_entropy)

y_pred_resample = clf_entropy.predict(X_test_res)
print ('accuracy_resample',accuracy_score(Y_test_res, y_pred_resample)*100)


y_hat_probs_resample = clf_entropy.predict_proba(X_test_res)[:, 1:]
print('auc_resample',roc_auc_score(Y_test_res, y_hat_probs_resample)*100)

confusion_matrix(Y_test_res, y_pred_resample)
```

###20

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=3, n_estimators=5)
clf.fit(X_train_res,Y_train_res)

y_pred_resample = clf.predict(X_test_res)
print ('accuracy:',accuracy_score(Y_test_res, y_pred_resample)*100)


y_hat_probs_resample = clf.predict_proba(X_test_res)[:, 1:]
print('auc:',roc_auc_score(Y_test_res, y_hat_probs_resample)*100)

#Yes, it's better, more accurate 
```
